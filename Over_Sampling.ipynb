```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import ADASYN
```

```python
Dados = pd.read_csv('Dados_Simulados.csv')

X = Dados[['Renda', 'Exposição']]

y = Dados['Uso']
```

```python
ROS_X, ROS_y = RandomOverSampler().fit_resample(X, y)

ROS = ROS_X.join(ROS_y)

plt.figure(figsize = (15, 5))

Grafico_RUS = sns.scatterplot(data = ROS, x = 'Renda', y = 'Exposição', hue = 'Uso', s = 75)

Grafico_RUS.spines['top'].set_visible(False)

Grafico_RUS.spines['right'].set_visible(False)

plt.title('Random Over-Sampler')

plt.show()
```

```python
SMOTE_X, SMOTE_y = SMOTE().fit_resample(X, y)

SMOTE = SMOTE_X.join(SMOTE_y)

plt.figure(figsize = (15, 5))

Grafico_RUS = sns.scatterplot(data = SMOTE, x = 'Renda', y = 'Exposição', hue = 'Uso', s = 75)

Grafico_RUS.spines['top'].set_visible(False)

Grafico_RUS.spines['right'].set_visible(False)

plt.title('SMOTE')

plt.show()
```

```python
ADASYN_X, ADASYN_y = ADASYN().fit_resample(X, y)

ADASYN = ADASYN_X.join(ADASYN_y)

plt.figure(figsize = (15, 5))

Grafico_RUS = sns.scatterplot(data = ADASYN, x = 'Renda', y = 'Exposição', hue = 'Uso', s = 75)

Grafico_RUS.spines['top'].set_visible(False)

Grafico_RUS.spines['right'].set_visible(False)

plt.title('ADASYN')

plt.show()
```
