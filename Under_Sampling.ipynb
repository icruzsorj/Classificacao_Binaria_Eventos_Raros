```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from imblearn.under_sampling import RandomUnderSampler
from imblearn.under_sampling import NearMiss
```

```python
Dados = pd.read_csv('Dados_Simulados.csv')

Dados.head()

X = Dados[['Renda', 'Exposição']]

y = Dados['Uso']
```

```python
RUS_X, RUS_y = RandomUnderSampler().fit_resample(X, y)

RUS = RUS_X.join(RUS_y)

plt.figure(figsize = (15, 5))

sns.scatterplot(data = RUS, x = 'Renda', y = 'Exposição', hue = 'Uso', s = 75)

plt.title('Random Under-Sampler')

plt.show()
```

```python
NearMiss_1_X, NearMiss_1_y = NearMiss(version = 1, n_neighbors = 3).fit_resample(X, y)

NearMiss_1 = NearMiss_1_X.join(NearMiss_1_y)

plt.figure(figsize = (15, 5))

sns.scatterplot(data = NearMiss_1, x = 'Renda', y = 'Exposição', hue = 'Uso', s = 75)

plt.title('Near Miss 1')

plt.show()
```

```python
NearMiss_2_X, NearMiss_2_y = NearMiss(version = 2, n_neighbors = 3).fit_resample(X, y)

NearMiss_2 = NearMiss_2_X.join(NearMiss_2_y)

plt.figure(figsize = (15, 5))

sns.scatterplot(data = NearMiss_2, x = 'Renda', y = 'Exposição', hue = 'Uso', s = 75)

plt.title('Near Miss 2')

plt.show()
```

```python
NearMiss_3_X, NearMiss_3_y = NearMiss(version = 3, n_neighbors_ver3 = 5).fit_resample(X, y)

NearMiss_3 = NearMiss_3_X.join(NearMiss_3_y)

plt.figure(figsize = (15, 5))

sns.scatterplot(data = NearMiss_3, x = 'Renda', y = 'Exposição', hue = 'Uso', s = 75)

plt.title('Near Miss 3')

plt.show()
```
