```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
```

```python
np.random.seed(0)
```

```python
n = 10000

X, y = make_classification(n_samples = n, n_features = 2, n_redundant = 0, weights = [0.95])

Dados = pd.DataFrame(np.hstack((X, y.reshape(-1, 1))), columns = ['Renda', 'Exposição', 'Uso'])

Dados.head()
```

```python
plt.figure(figsize = (15, 5))

sns.scatterplot(data = Dados, x = 'Renda', y = 'Exposição', hue = 'Uso', s = 75)

plt.show()
```

```python
Dados.to_csv('Dados_Simulados.csv', index = False)
```
